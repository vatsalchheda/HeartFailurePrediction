```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
df = pd.read_csv("heart_failure_clinical_records_dataset.csv")
```

```python
df.info()
```

```python
plt.figure(figsize=(14,5))
sns.heatmap(df.corr(),annot = True,linewidth=0.5)
```

```python
df.columns
```

```python
features = ['age', 'anaemia', 'creatinine_phosphokinase', 'diabetes','ejection_fraction', 'high_blood_pressure', 'platelets','serum_creatinine', 'serum_sodium', 'sex', 'smoking', 'time']
```

```python
X = df[features]
```

```python
y = df['DEATH_EVENT']
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
```

```python
train_x,test_x,train_y,test_y = train_test_split(X,y,test_size=0.2,train_size=0.8,random_state=1)
```

```python
sc = StandardScaler()
```

```python
train_x = sc.fit_transform(train_x)
```

```python
test_x = sc.transform(test_x)
```

```python
from sklearn.svm import SVC
classifier = SVC(kernel = 'linear', random_state = 0)
classifier.fit(train_x, train_y)
```

```python
y_pred = classifier.predict(test_x)
```

```python
len(y_pred)
```

```python
test_y.count()
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score
cm = confusion_matrix(test_y, y_pred)
print(cm)
accuracy_score(test_y, y_pred)
```
